# sanitizing specific columns

```python
# Handling Missing Values
# For example, impute missing values in the 'Age' column with the mean
age_mean = titanic_data['Age'].mean()
titanic_data['Age'].fillna(age_mean, inplace=True)

# Converting Data Types
# For example, convert the 'Fare' column to float
titanic_data['Fare'] = titanic_data['Fare'].astype(float)

# Removing Outliers
# For example, cap outliers in the 'Fare' column at a certain threshold
fare_cap = 300  # Example threshold
titanic_data['Fare'] = titanic_data['Fare'].apply(lambda x: min(x, fare_cap))

# Removing Leading/Trailing Whitespaces
# For example, remove leading/trailing whitespaces from the 'Sex' column
titanic_data['Sex'] = titanic_data['Sex'].str.strip()

# Display the first few rows of the dataset after sanitizing specific columns
print("First few rows of the Titanic dataset after sanitizing specific columns:")
print(titanic_data.head())
```

# eliminate the data sets NaN values

```python
# Drop rows with NaN values
titanic_data.dropna(inplace=True)

# Check the number of rows after removing NaN values
num_rows_after = titanic_data.shape[0]
print("Number of rows after removing NaN values:", num_rows_after)
```

```python
import pandas as pd

# Load Titanic dataset
titanic_data = pd.read_csv('titanic.csv')

# Create a new feature 'Family_Size' by summing 'SibSp' and 'Parch'
titanic_data['Family_Size'] = titanic_data['SibSp'] + titanic_data['Parch']

# Create a new feature 'Traveling_Alone' indicating whether a passenger is traveling alone or with family
titanic_data['Traveling_Alone'] = (titanic_data['Family_Size'] == 0).astype(int)

# Display the first few rows of the dataset with the new features
print("First few rows of the Titanic dataset with new features:")
print(titanic_data.head())
```
